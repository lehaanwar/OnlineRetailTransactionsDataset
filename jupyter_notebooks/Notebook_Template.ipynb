```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#load the cleaned dataset
df_cleaned = pd.read_csv('../dataset/Cleandataset/Cleandataset.csv')
print(f" Loaded cleaned data: {df_cleaned.shape[0]:,} rows  {df_cleaned.shape[1]} columns")
df_cleaned.head()
```

```python
if 'TotalTransactionValue' not in df_cleaned.columns:
    df_cleaned['TotalTransactionValue'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
df_cleaned.head()


```

First Hypothesis - Certain countries have higher total transaction values due to larger customer bases or higher purchasing power.

```python
#Show heatmap of total transaction values by country.
plt.figure(figsize=(12, 8))
country_totals = df_cleaned.groupby('Country')['TotalTransactionValue'].sum().reset_index()
country_pivot = country_totals.pivot_table(values='TotalTransactionValue', index='Country')
sns.heatmap(country_pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Total Transaction Values by Country')
plt.show()
```

```python
# show histogram of total transaction values by country 
plt.figure(figsize=(12, 8))
country_totals = df_cleaned.groupby('Country')['TotalTransactionValue'].sum().reset_index()
sns.barplot(data=country_totals, x='Country', y='TotalTransactionValue')
plt.title('Total Transaction Values by Country')
plt.show()
```

```python
# calculate total sales by country
country_sales = df_cleaned.groupby('Country')['TotalTransactionValue'].sum().reset_index()
country_sales.head()

```

```python
#sort country by total sales in descending order
country_sales = country_sales.sort_values(by='TotalTransactionValue', ascending=False)
country_sales.head()
```

```python
#sort country sales in ascending order for better visualization
country_sales = country_sales.sort_values(by='TotalTransactionValue', ascending=True)
country_sales.head()


```

```python
#bar chart of sales in descending order
plt.figure(figsize=(12, 8))
sns.barplot(data=country_sales, x='Country', y='TotalTransactionValue', palette='viridis')
plt.title('Total Sales by Country')
plt.show()
```

Shows different countries have higher sales values than others. It could be because of larger customer bases but I would need more information to really compare. So that hypothesis is not shown very clearly.

```python
#load the cleaned dataset
df_cleaned = pd.read_csv('../dataset/Cleandataset/Cleandataset.csv')
print(f" Loaded cleaned data: {df_cleaned.shape[0]:,} rows  {df_cleaned.shape[1]} columns")
df_cleaned.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

Second Hypothesis - Products that have higher unit prices due to demand spikes during specific times of the year.

```python
#Use TimeSeries for time-based analysis
df_cleaned['InvoiceDate'] = pd.to_datetime(df_cleaned['InvoiceDate'])
df_cleaned.head()

```

```python
#Time component extraction
df_cleaned['Year'] = df_cleaned['InvoiceDate'].dt.year
df_cleaned['Month'] = df_cleaned['InvoiceDate'].dt.month
df_cleaned['Day'] = df_cleaned['InvoiceDate'].dt.day    
df_cleaned['MonthName'] = df_cleaned['InvoiceDate'].dt.month_name()
df_cleaned.head()



```

```python
#Filter for seasonal months (e.g., November and December, January, February)
seasonal_months = seasonal_months = df_cleaned[df_cleaned['Month'].isin([11, 12, 1, 2])]
df_seasonal = df_cleaned[df_cleaned['Month'].isin(seasonal_months)]
df_seasonal.head()
```

```python
if 'TotalTransactionValue' not in df_cleaned.columns:
    df_cleaned['TotalTransactionValue'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
df_cleaned.head()
```

```python
# Calculate total sales by month to analyze seasonal trends
monthly_sales = seasonal_months.groupby('MonthName')['TotalTransactionValue'].sum().reset_index()
monthly_sales.head()

```

The above shows that November and December have higher sales than January and February.

```python
#Monthly sales bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=monthly_sales, x='MonthName', y='TotalTransactionValue', order=['January', 'February', 'November', 'December'])
plt.title('Monthly Sales in Seasonal Months')
plt.show()
```

```python
#Monthly sales line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='MonthName', y='TotalTransactionValue', marker='o', sort=False)
plt.title('Monthly Sales in Seasonal Months')
plt.show()
```

Third Hypothesis - Lower priced products will encourage higher quantities purchased.

```python
#show scatter plot of Quantity vs UnitPrice to see if there's any correlation.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='UnitPrice', y='Quantity')
plt.title('Scatter Plot of Quantity vs UnitPrice')
plt.show()
```

```python
#heatmap of Quantity vs UnitPrice
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned.pivot_table(values='Quantity', index='UnitPrice'), cmap='YlGnBu')
plt.title('Heatmap of Quantity vs UnitPrice')
plt.show()
```

The above plots show the lower the Unit Price, the higher the Quantity of products purchased.
