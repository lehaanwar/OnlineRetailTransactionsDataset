```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load the raw dataset
df = pd.read_csv('../dataset/Rawdata/Online-Retail.csv')
df.head(10)
```

```python
# handle missing values
# check missing values in the 'Description' column
df_missing = df[df['Description'].isnull()]
df_missing.head(10)
```

```python
# check if missing values correlate with 'country' column
df_missing['Country'].value_counts()
# all the missing values are from 'United Kingdom' - these could be due to data entry errors in the UK warehouse or a system glitch affecting only UK entries. Further investigation is needed to determine the exact cause.
```

```python
#Remove rows with missing values in the 'Description' column
df = df.dropna(subset=['Description'])

# Check again for missing values
df.isnull().sum()



```

```python
#visual representation of missing values
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

```

```python
# save the cleaned dataset
df.to_csv('Cleandataset.csv', index=False)
```

```python
import pandas as pd

#load cleaned dataset
df_cleaned = pd.read_csv('../dataset/Cleandataset/Cleandataset.csv')
print(f" Loaded cleaned data: {df_cleaned.shape[0]:,} rows  {df_cleaned.shape[1]} columns")

# Quick preview
df_cleaned.head()
```

```python
#check for outliers
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned, x='Quantity')
plt.title('Boxplot of Quantity')
plt.show()
```

```python
# show outliers in 'Quantity' column
df_cleaned['Quantity'].describe()
```

```python
#Remove retail transactions with negative 'Quantity' values
df_cleaned = df_cleaned[df_cleaned['Quantity'] >= 0]
#Check the statistics again
df_cleaned['Quantity'].describe()
```

```python
#As there is a high maximum value in 'Quantity' column, we will further investigate transactions with Quantity greater than 1000
df_high_quantity = df_cleaned[df_cleaned['Quantity'] > 1000]
df_high_quantity
```

```python
#Remove high outliers in 'Quantity' column by setting a threshold of 1000
df_cleaned = df_cleaned[df_cleaned['Quantity'] <= 1000]
#Check the statistics again
df_cleaned['Quantity'].describe()
```

---

```python
# show outliers in 'UnitPrice' column
df_cleaned['UnitPrice'].describe()
```

```python
#Remove outliers in negative 'UnitPrice' values
df_cleaned = df_cleaned[df_cleaned['UnitPrice'] >= 0]   
#check the statistics again
df_cleaned['UnitPrice'].describe()

```

```python
# Remove outliers in high 'UnitPrice' values by setting a threshold of 100
df_cleaned = df_cleaned[df_cleaned['UnitPrice'] <= 100]
#check the statistics again
df_cleaned['UnitPrice'].describe()
```
